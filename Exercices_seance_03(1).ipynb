```python
import re

def compte_mots(list):
    frequence = {} #we set the frequence to 0
    for mot in list:
        if mot in frequence:
            frequence[mot] += 1
        else:
            frequence[mot] = 1
    return (frequence)  

def sauvegarde(most_frequent, least_frequent):
    file1 = open("mots_peu_frequents.txt", "w")
    file2 = open("mots_très_frequents.txt", "w")
 
    file1.write(" ".join(least_frequent))
    file2.write(" ".join(most_frequent))

    file1.close()
    file2.close()


def main(path, N, M = 10):

    with open(path, "r", encoding="utf-8") as f: #reading of the file in path, reads french accents
        texte = f.read().lower() #lowercase text
    
    mots = re.findall(r"\b\w+\b", texte, flags=re.UNICODE) #r"\b\w+\b" = raw"limit\alphanumerical chain of characters\limit"
    frequence = compte_mots(mots) #count frequence of all words of the text

    increasing_frequence_order = sorted(frequence.items(), key=lambda kv: kv[1]) #kv[1] = frequence of the word contained in the tuple kv, key=lambda ->take kv[1] value (frequence)

    least_frequent = [mot for mot, nb in increasing_frequence_order[:M]]

    decreasing_frequence_order = sorted(frequence.items(), key=lambda kv: kv[1], reverse=True)
    
    most_frequent = [mot for mot, nb in decreasing_frequence_order[:N]]

    sauvegarde(most_frequent, least_frequent)
 
    return (most_frequent, least_frequent)


if __name__ == "__main__":
    main("./corpus/1912_Loti-Pierre_Un-pelerin-d-Angkor.txt" , 10, 10)
```
